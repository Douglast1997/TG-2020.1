```python
def test_lightgbm_cpu():
    import numpy as np
    import pandas as pd
    import lightgbm as lgb

    X1 = np.repeat(np.arange(10), 1000)
    X2 = np.repeat(np.arange(10), 1000)
    np.random.shuffle(X2)

    y = (X1 + np.random.randn(10000)) * (X2 + np.random.randn(10000))
    data = pd.DataFrame({'y': y, 'X1': X1, 'X2': X2})

    lgb_params = {'learning_rate': 0.1,
                  'boosting': 'gbdt',
                  'objective': 'regression',
                  'metric': 'rmse',
                  'feature_fraction': 0.9,
                  'bagging_fraction': 0.75,
                  'num_leaves': 31,
                  'bagging_freq': 1,
                  'min_data_per_leaf': 250}
    lgb_train = lgb.Dataset(data=data[['X1', 'X2']], label=data.y)
    cv = lgb.cv(lgb_params,
                lgb_train,
                num_boost_round=100,
                early_stopping_rounds=15,
                stratified=False,
                verbose_eval=50) 
```

# K-fold SVR:

```python
import pandas
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.svm import SVR
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("databasesTCC/Outdoordf.csv")
#dataset
```

```python
X = dataset.iloc[:, 12:]
z = dataset.iloc[:, 0]
y_names = ['rssi_1_1','rssi_1_2','rssi_1_3','rssi_2_1','rssi_2_2','rssi_2_3','rssi_3_1','rssi_3_2','rssi_3_3']
```

```python
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)
scores = []
best_svr = SVR(C=64, cache_size=200, coef0=0.1, degree=3, epsilon=0.5, gamma=8, kernel='rbf', max_iter=-1, shrinking=True, tol=0.001, verbose=False)
cv = KFold(n_splits=10, random_state=42, shuffle=True)
y = dataset.iloc[:, 1]
#y.head()
```

```python
y_columns_names = []
y_names = ['rssi_1_1','rssi_1_2','rssi_1_3','rssi_2_1','rssi_2_2','rssi_2_3','rssi_3_1','rssi_3_2','rssi_3_3']
columns = ['_Fold:1', '_Fold:2', '_Fold:3', '_Fold:4', '_Fold:5','_Fold:6','_Fold:7','_Fold:8','_Fold:9','_Fold:10']
i = 0
for label in y_names:
    #print(label)
    for fold in columns:
        #print(fold)
        #print(label + fold)
        y_columns_names.append(label + fold)
        i = i + 1
#y_columns_names
```

```python
df_y_pred = pd.DataFrame(index = range(0,5986), columns= y_columns_names)
df_y_pred = df_y_pred.fillna(0) # with 0s rather than NaNs
df_y_pred.head()
```

```python
for i in range(1, 10):
    y = dataset.iloc[:, i]
    j = 1
    for train_index, test_index in cv.split(X):
        X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
        train_index
        z_index = z[test_index]
        best_svr.fit(X_train, y_train)
        #scores.append(best_svr.score(X_test, y_test))#Somar mais 600 dados indoor
        df_meds_CDB = pd.read_csv("./CDB/CDB_20.csv") #X_test
        df_meds_CDB = df_meds_CDB.drop("lat", axis = 1)
        df_meds_CDB = df_meds_CDB.drop("lon", axis = 1)
        X_test = df_meds_CDB
        y_pred = best_svr.predict(X_test)
        df_y_pred[y.name+'_Fold:'+str(j)] = pd.DataFrame({y.name: y_pred})
        j = j + 1
```

```python
df_y_pred.head()
```

```python
k = 10 #Number of k-fold

#df_y_pred_rssi_1_1:
df_y_pred_rssi_1_1 = 0
for i in range(1,k+1):
    df_y_pred_rssi_1_1 =  df_y_pred_rssi_1_1 + df_y_pred['rssi_1_1_Fold:'+ str(i)]
df_y_pred_rssi_1_1 = df_y_pred_rssi_1_1/k
#df_y_pred_rssi_1_1.head()

#df_y_pred_rssi_1_2:
df_y_pred_rssi_1_2 = 0
for i in range(1,k+1):
    df_y_pred_rssi_1_2 =  df_y_pred_rssi_1_2 + df_y_pred['rssi_1_2_Fold:'+ str(i)]
df_y_pred_rssi_1_2 = df_y_pred_rssi_1_2/k
#df_y_pred_rssi_1_2.head()

#df_y_pred_rssi_1_3:
df_y_pred_rssi_1_3 = 0
for i in range(1,k+1):
    df_y_pred_rssi_1_3 =  df_y_pred_rssi_1_3 + df_y_pred['rssi_1_3_Fold:'+ str(i)]
df_y_pred_rssi_1_3 = df_y_pred_rssi_1_3/k
#df_y_pred_rssi_1_3.head()

#df_y_pred_rssi_2_1:
df_y_pred_rssi_2_1 = 0
for i in range(1,k+1):
    df_y_pred_rssi_2_1 =  df_y_pred_rssi_2_1 + df_y_pred['rssi_2_1_Fold:'+ str(i)]
df_y_pred_rssi_2_1 = df_y_pred_rssi_2_1/k
#df_y_pred_rssi_2_1.head()

#df_y_pred_rssi_2_2:
df_y_pred_rssi_2_2 = 0
for i in range(1,k+1):
    df_y_pred_rssi_2_2 =  df_y_pred_rssi_2_2 + df_y_pred['rssi_2_2_Fold:'+ str(i)]
df_y_pred_rssi_2_2 = df_y_pred_rssi_2_2/k
#df_y_pred_rssi_2_2.head()

#df_y_pred_rssi_2_3:
df_y_pred_rssi_2_3 = 0
for i in range(1,k+1):
    df_y_pred_rssi_2_3 =  df_y_pred_rssi_2_3 + df_y_pred['rssi_2_3_Fold:'+ str(i)]
df_y_pred_rssi_2_3 = df_y_pred_rssi_2_3/k
#df_y_pred_rssi_2_3.head()

#df_y_pred_rssi_3_1:
df_y_pred_rssi_3_1 = 0
for i in range(1,k+1):
    df_y_pred_rssi_3_1 =  df_y_pred_rssi_3_1 + df_y_pred['rssi_3_1_Fold:'+ str(i)]
df_y_pred_rssi_3_1 = df_y_pred_rssi_3_1/k
#df_y_pred_rssi_3_1.head()

#df_y_pred_rssi_3_2:
df_y_pred_rssi_3_2 = 0
for i in range(1,k+1):
    df_y_pred_rssi_3_2 =  df_y_pred_rssi_3_2 + df_y_pred['rssi_3_2_Fold:'+ str(i)]
df_y_pred_rssi_3_2 = df_y_pred_rssi_3_2/k
#df_y_pred_rssi_3_2.head()

#df_y_pred_rssi_3_3:
df_y_pred_rssi_3_3 = 0
for i in range(1,k+1):
    df_y_pred_rssi_3_3 =  df_y_pred_rssi_3_3 + df_y_pred['rssi_3_3_Fold:'+ str(i)]
df_y_pred_rssi_3_3 = df_y_pred_rssi_3_3/k
#df_y_pred_rssi_3_3.head()    
```
